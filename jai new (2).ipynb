```python
from jupyter_dash import JupyterDash
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic
```

```python
df = pd.read_csv("Statewise_General_Index_july2019_20Aug2020_dec20_2_5.csv")
df.head(15)
```

```python
# Initialize the app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    
    html.H1("Statewise_General_Index of 2020-2022"),
    
    # Histogram
    dcc.Graph(
        id='histogram',
        figure={
            'data': [{
                'x': df['Tamil Nadu'],
                'type': 'histogram'
            }],
            'layout': {
                'title': 'Histogram',
                 'xaxis' : {'title' : 'Tamil Nadu'}
               
            }
            }
        
    ),
    
    # Box plot
    dcc.Graph(
        id='boxplot',
        figure={
            'data': [{
                'x': df['Year'],
                'y': df['Andhra Pradesh'],
                'type': 'box'
            }],
            'layout': {
                'title': 'Box plot',
                'xaxis' : {'title' : 'Year'},
                'yaxis' : {'title' : 'Andhra Pradesh'}
                
            }
        }
    ),
    # Mosaic plot
    dcc.Graph(
        id='mosaicplot',
        figure={
            'data': [{
                'y': df['Goa'],
                'type': 'mosaicplot'
            }],
            'layout': {
                'title': 'Mosaic plot',
                'yaxis' : {'title' : 'Goa'}
            }
        }
    ),
    
    # Scatter plot
    dcc.Graph(
        id='scatterplot',
        figure={
            'data': [{
                'x': df['Andhra Pradesh'],
                'y': df['Tamil Nadu'],
                'mode': 'Growth',
                'marker': {
                    'size': 15,
                    'color': 'blue',
                    'opacity': 0.7
                }
            }],
            'layout': {
                'title': 'Scatter plot',
                'xaxis' : {'title' : 'Andhra Pradesh'},
                'yaxis' : {'title' : 'Tamil Nadu'}
                
            }
        }
    ),
    
    # Bar chart
    dcc.Graph(
        id='barchart',
        figure={
            'data': [{
                'x': df['Year'],
                'y': df['Puducherry'],
                'type': 'bar'
            }],
            'layout': {
                'title': 'Bar chart',
                'xaxis' : {'title' : 'Year'},
                'yaxis' : {'title' : 'Puducherry'}
                
                
            }
        }
    )
])
```

```python
# Run the app
if __name__ == '__main__':
    app.run_server(port=2050)
```
